```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('/content/dataset.cv')
dataset
```

```python
X = dataset.iloc[:, [2, 3]].values
y = dataset.iloc[:, 4].values
display(X)
```

```python
x=df["Age"].mode()
x
```

```python
y=df["Estimated Salary"].median()
y
```

```python
#splitting the dataset into training and test sets
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test,= train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
st_x= StandardScaler()
X_train=st_x.fit_transform(X_train)
X_test=st_x.transform(X_test)
X_test
```

Fitting K-NN classifier to the Training data : Now we will fit the K-NN classifier to thge training data. To do this we will import the KNeighboursClassifier class of Sklearn Neighbors library. After importing the class, we will create the Classifier object of the class. The Parameter of thisÂ classÂ willÂ be

    n_neighbors : To define the required neighbors of the algorithm. Usually, it takes 5.
    metric = 'minkwski' : This is the default parameter and it decides the distance between the points.
    p = 2 : It is equivalent to the standard Euclidean metric.

```python
# Fitting KNN classifier to the training set
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

Predicting the Test Result : To predict the test set result, we will create a y_pred vector as we did in Logistic Regression. Below is theÂ codeÂ forÂ itÂ :

```python
y_pred = classifier.predict(X_test)
y_pred
```
